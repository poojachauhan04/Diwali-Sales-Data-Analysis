```python
# Importing essential libraries for data analysis and visualization.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset and check its dimensions (rows, columns)
df=pd.read_csv(r'C:\Users\hp\OneDrive\Documents\DA Course\Python_Learning\Python Projects\Diwali_Sales_Analysis\Diwali Sales Data.csv',encoding='unicode_escape')
df.shape
```

```python
# Display the first 5 rows of the dataset to preview data.
df.head()
```

```python
# Display dataset summary: rows, columns, non-null counts, data types, and memory usage.
df.info()
```

```python
# Drop unrelated/blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
# Check dataset dimensions (rows, columns)
df.shape
```

```python
# Check total missing values in each column
df.isnull().sum()
```

```python
# Replace missing values in 'Amount' column with its mean
df.fillna(df['Amount'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# Count total number of duplicate rows in the DataFrame
df.duplicated().sum()
```

```python
# Dropping duplicate from dataset.
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Change Data Type.
df['Amount']=df['Amount'].astype('int')
```

```python
# Check the data type of 'Amount' column
df['Amount'].dtype
```

```python
# List all column names in the dataset
df.columns
```

```python
# Rename column name.
df.rename(columns={'Orders':'Total Orders'},inplace=True)
```

```python
# Generate summary statistics (count, mean, std, min, quartiles, max) for numeric columns
df.describe()
```

```python
# Generate summary statistics only for Age, Total Orders, and Amount columns
df[['Age','Total Orders','Amount']].describe()
```

****Exploratory Data Analysis****

**Gender :-**

```python
# Create a count plot of customers by Gender with labels on bars.
plt.figure(figsize=(4,6))
x=sns.countplot(data=df,x='Gender',hue='Gender', palette=['#FF69B4','blue'],width=0.4)
# Add labels on top of each bar
for bars in x.containers:
    x.bar_label(bars)
plt.title('Gender-Wise Count',fontsize=12,fontweight='bold')
plt.show()
```

```python
# Group sales by Gender and sum Amount.
Gender_Wise_Sales=df.groupby('Gender')['Amount'].sum().sort_values(ascending=False).reset_index()
Gender_Wise_Sales
```

```python
# Plot bar chart of Gender-wise Sales Amount.
plt.figure(figsize=(4,6))
y=sns.barplot(data=Gender_Wise_Sales,x='Gender',y='Amount',hue='Gender', palette=['#FF69B4','blue'],width=0.4)
plt.title('Gender-wise Sales Amount',fontsize=12,fontweight='bold')
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.

**Age :-**

```python
# Create a count plot of Age Groups split by Gender with labels on bars.
plt.figure(figsize=(10,4))
x=sns.countplot(data=df,y='Age Group',hue='Gender', palette=['#FF69B4','blue'])
for bars in x.containers:
    x.bar_label(bars)
plt.title('Gender-Wise Count',fontsize=12,fontweight='bold')
plt.show()
```

```python
#Sales by Age Group, sum the Amount, and sort in descending order
AgeGroup_Wise_Amount=df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False).reset_index()
```

```python
# Plot total Amount spent by each Age Group as a horizontal bar chart.
plt.figure(figsize=(10,4))
sns.barplot(data=AgeGroup_Wise_Amount,x='Amount',y='Age Group',color='tan')
plt.title('AgeGroup-Wise Amount Spent',fontsize=12,fontweight='bold')
plt.show()
```

From above graphs we can see that most of the buyers are of age-group between 26-35 years female.

**State :-**

```python
# Total number of orders from top 10 states.
df.columns
```

```python
# Find top 10 states by total orders and sort in descending order
Top_States_by_order=df.groupby('State')['Total Orders'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
# Plot top 10 states by total orders as a vertical bar chart
sns.set_theme(rc={'figure.figsize':(8,3)})
sns.barplot(data=Top_States_by_order,x='State',y='Total Orders',color='purple')
plt.title('Top 10 States by Order',fontsize=15,fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total Sales from top 10 states.
Top_States_by_amount=df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
# Plot top 10 states by total sales amount as a vertical bar chart.
sns.set_theme(rc={'figure.figsize':(8,3)})
sns.barplot(data=Top_States_by_amount,x='State',y='Amount',color='grey',width=0.6)
plt.title('Top 10 States by Amount',fontsize=15,fontweight='bold')
plt.xticks(rotation=45)
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

From above graphs we can see that most of the orders and amound are from Uttar Pradesh, Maharashtra and Karnataka respectively.

**Marital Status :-**

```python
# Plot count of customers by Marital Status split by Gender.
sns.set_theme(rc={'figure.figsize':(5,3)})
x=sns.countplot(data=df,x='Marital_Status',hue='Gender',palette=['#FF69B4','blue'],width=0.4)
plt.title('Total Customers by Marital Status',fontsize=12,fontweight='bold')
for bars in x.containers:
    x.bar_label(bars)
plt.show()
```

**Occupation :-**

```python
# Plot count of customers by Occupation with labels on bars
sns.set_theme(rc={'figure.figsize':(8,3)})
x=sns.countplot(data=df,x='Occupation',color='skyblue')
plt.title('Total Customers by Occupation',fontsize=12,fontweight='bold')
# Rotate x-axis labels for readability
plt.xticks(rotation=90)
# Add labels on top of each bar
for bars in x.containers:
    x.bar_label(bars)
plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Aviation and Healthcare sector.
